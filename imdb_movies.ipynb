```python
#question4
import pandas as pd

df = pd.read_csv('C:/Users/agent/Desktop/imdb_movies.csv')

df['runtime'] = df['runtime'].astype(int)
df = df[df['runtime'] != 0]
df['runtime_hours'] = df['runtime'] // 60
df['runtime_minutes'] = df['runtime'] % 60

average_runtime_per_year = df.groupby('release_year')['runtime'].mean()

for year, runtime in average_runtime_per_year.items():
    hours = int(runtime // 60)
    minutes = int(runtime % 60)
    print(f"The average length of the movie of the year{year}: {hours} hour and {minutes} min")
```

```python
#question6
import pandas as pd
df = pd.read_csv('C:/Users/agent/Desktop/imdb_movies.csv')
df['runtime'] = df['runtime'].astype(int)
df = df[df['runtime'] != 0]
df['runtime_hours'] = df['runtime'] // 60
df['runtime_minutes'] = df['runtime'] % 60
longest_movies_per_year = df.groupby('release_year').apply(lambda x: x[x['runtime'] == x['runtime'].max()])
shortest_movies_per_year = df.groupby('release_year').apply(lambda x: x[x['runtime'] == x['runtime'].min()])
print("The longest running movies for each year:")
for year, movie in longest_movies_per_year[['release_year', 'original_title', 'runtime_hours', 'runtime_minutes']].iterrows():
    print(f"Year: {year},title: {movie['original_title']}, runtime: {movie['runtime_hours']} hour and {movie['runtime_minutes']} min")
print("\n The shortest running movies for each year:")
for year, movie in shortest_movies_per_year[['release_year', 'original_title', 'runtime_hours', 'runtime_minutes']].iterrows():
    print(f"Year: {year}, title: {movie['original_title']}, runtime: {movie['runtime_hours']} hour and {movie['runtime_minutes']} min")
```

```python
#question13
import pandas as pd

df = pd.read_csv('C:/Users/agent/Desktop/imdb_movies.csv')
df = df[df['revenue_adj'] != 0]
df = df[df['budget_adj'] != 0]
df['revenue_adj'] = df['revenue_adj'].replace(',', '').astype(float)
df['budget_adj'] = df['budget_adj'].replace(',', '').astype(float)

df['profit'] = df['revenue_adj'] - df['budget_adj']

most_profitable_movie = df[df['profit'] == df['profit'].max()]

print("The most profitable movie is Ice Tar:")
print(f"title: {most_profitable_movie['original_title'].values[0]}")
print(f"Profit: ${most_profitable_movie['profit'].values[0]:,.2f} Million")
```
